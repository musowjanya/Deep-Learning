```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
import pandas as pd
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
dt=load_iris()
```

```python
dt.keys()
```

```python
X,Y=dt.data,dt.target
```

```python
dt.target_names
```

```python
model=nn.Sequential(
    nn.Linear(4,30),
    nn.ReLU(),
    nn.Linear(30,10),
    nn.ReLU(),
    nn.Linear(10,3),
    # nn.Softmax(dim=1)
)
```

```python
X=torch.FloatTensor(X)
Y=torch.LongTensor(Y)
```

```python
# model(X) #output is three colomns because of three target classes
```

```python
# torch.argmax(model(X),dim=1)
```

```python
# lossfn=nn.LogSoftmax()
lossfn=nn.CrossEntropyLoss()
optsgd=optim.SGD(model.parameters(),lr=0.01)
```

Training

```python
for _ in range(100):
  yp=model(X)
  loss=lossfn(yp,Y)
  loss.backward()
  optsgd.step()

  print(loss.item())
```

Inference

```python
ynewp=model(X)
```

```python
sfout=torch.softmax(ynewp,dim=1)
```

```python
torch.argmax(sfout,dim=1)
```

Using Batches

```python
from torch.utils.data import DataLoader
```

```python
df=pd.DataFrame(X)
```

```python
df
```

```python
df['Y']=Y
```

```python
df=torch.tensor(df.numpy())
```

```python
model = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
lossfn = nn.CrossEntropyLoss()
batchSgd = optim.SGD(model.parameters(), lr=0.01)
```

```python
batches=DataLoader(df,batch_size=10, shuffle=True)
```

```python
# for batch in batches:
#   X=batch[:,:-1]
#   Y=batch[:,-1]

#   print(X,Y)
```

```python
model.train()
for epoch in range(200):
  batch_loss=0
  for i,batch in enumerate(batches):
    batch_X=batch[:,:-1].float()
    batch_Y=batch[:,-1].long()

    batchSgd.zero_grad()
    batch_yp=model(batch_X)
    loss=lossfn(batch_yp,batch_Y)
    loss.backward()
    batchSgd.step()
    batch_loss+=loss.item()
  print(f"Epoch {epoch+1}: Total Loss={batch_loss}")
```
